#### Task 1

For the three natural numbers entered (user enters numbers through the space), find the minimum and maximum

```python
numbers = input('Enter three numbers in space: ')
tape_numbers=numbers.split()       # I translate the tape into a list
new_numbers = []                  # Create a new list and add items from the list of tapes in int format to it
for i in tape_numbers:
    new_numbers.append(int(i))

min_number = min(new_numbers)
print('Minimum number:', min_number)
print('='*20)
max_number = max(new_numbers)
print('Maximum number:', max_number)
```
