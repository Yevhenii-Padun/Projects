```python
news = []
for el in topic_5_name:
    topic_name=el.text
    href_topic_ = el.get('href')
    href_topic = link + href_topic_
    link2 = href_topic
        
    r = requests.get(link2, headers=headers)
    soup2 = BeautifulSoup(r.text, 'lxml')
    tags = soup2.find_all(class_="article article_list")
    
    for i in range(len(tags)):
        a_tags = tags[i].find('a').text
        href_tags_ = tags[i].find('a').get('href')
        href_tags = link+href_tags_
        data2 = tags[i].find(class_="article_author").text
        e=data2.split(' — ')
        data, time= e[0].split(', ')
        section = e[1]
        news.append({'topic_name': topic_name, 'href_topic': href_topic, 'news_name': a_tags,
                     'href_news': href_tags,  'data':  data, 'time':time, 'section':section })
news
```

```python
import json
```

```python
json.dumps(news, ensure_ascii=False)
```
