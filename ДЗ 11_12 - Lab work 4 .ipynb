2. Отримати курс долара за останній рік.

```python
import requests
from datetime import datetime, timedelta

YEAR_USD = []

date = datetime.now()
link = 'https://bank.gov.ua/NBU_Exchange/exchange_site?start=20210914&end=20220913&valcode=usd&sort=exchangedate&order=desc&json'
params = {
    'valcode' : 'USD', 
    'date' : date.strftime('%Y%m%d')
}

for i in range(1, 366):
    print(f'Опрацювали днів: {i}')
    try:
        result = requests.get(link, params).json()
        YEAR_USD.extend(result)
    except:
        print('Error')
    
    params['date'] = (date - timedelta(days=i)).strftime('%Y%m%d')
```

```python
YEAR_USD
```

```python
rates = [(day['exchangedate'], day['rate']) for day in YEAR_USD]
```

```python
rates[::-1]
```

3. Вивести середнє значення та відхилення курсу за кожний місяць.

```python
#створюю словник курсу долара за рік
excg_usd = {datetime.strptime(i.get('exchangedate'), '%d.%m.%Y') : i.get('rate_per_unit') for i in YEAR_USD}
print ({i.strftime('%d.%m.%Y'):excg_usd.get(i) for i in excg_usd.keys()})
```

```python
#Створюю словник, де ключами є місяці, а значеннями - курс доллара кожного дня в цьом місяці
excg_usd_for_month= {i : [excg_usd.get(y) for y in excg_usd.keys() if y.strftime('%Y-%m') == i]
                  for i in set(ii.strftime('%Y-%m') for ii in list(excg_usd.keys())) }
```

```python
#За допомогою бібліотеки statistics відсортовую по місяцям та виводжу середнє значення та відхтлення курсу
import statistics

print ('Місяць    ', 'Середнє значення     ', 'Відхилення урсу')
for el in sorted(excg_usd_for_month.keys()):
    print (el, '     ', '{:.4f}'.format(statistics.mean(excg_usd_for_month.get(el))),'            ', '{:.4f}'.format(statistics.stdev(excg_usd_for_month.get(el))))
```

4. Дану інформацію записати у файл за допомогою pickle.

```python
import pickle
pickle_file2 = {ii:[statistics.mean(excg_usd_for_month.get(ii)),statistics.stdev(excg_usd_for_month.get(ii))] 
                for ii in sorted(excg_usd_for_month.keys())}
   
print(pickle_file2)
```

```python
with open ('pickle_file2.pickle', 'wb') as f:
    pickle.dump (pickle_file2, f)
```
