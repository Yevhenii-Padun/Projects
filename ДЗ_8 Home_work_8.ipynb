```python
import csv
with open('covid_19_data.csv', 'r') as f:
    file_csv = csv.reader(f)
    data = list(file_csv)
```

```python
covid_dict = dict()
for i in range(len(data[0])):
    covid_dict[data[0][i]] = [row[i] for row in data[1:]]
```

```python
len(covid_dict)
```

```python
covid_dict.keys()
```

```python
sum_Country_Region = {key: sum(map(lambda i: int(i[0]), covid_dict['Country/Region'])) for key in covid_dict}
sum_Country_Region
```

```python
new_covid_dict = dict(zip(covid_dict['Country/Region'], covid_dict['Confirmed']))
new_covid_dict
```

```python
import pickle

with open('new_covid_dict.pickle2', mode='wb') as f:
    pickle.dump(new_covid_dict, f)
```
